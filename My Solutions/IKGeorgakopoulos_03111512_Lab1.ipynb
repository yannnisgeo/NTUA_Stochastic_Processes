# 1η Εργαστηριακή Άσκηση

Σε αυτή την άσκηση θα πραγματοποιήσουμε μία προσομοίωση του προβλήματος 5 του βιβλίου.

Στο σχήμα του βιβλίου φαίνεται η κάτοψη ενός σπιτιού με πέντε
δωμάτια: κουζίνα (Κ), βιβλιοθήκη (Β), σαλόνι (Σ), υπνοδωμάτιο (Υ) και μπάνιο
(Μ), καθώς και οι πόρτες που τα συνδέουν.  ́Ενα έντομο που ζει στο σπίτι κάθε
βράδυ διασχίζει τυχαία μία από τις πόρτες του δωματίου στο οποίο βρίσκεται
και παραμένει στο δωμάτιο που οδηγεί η πόρτα μέχρι το επόμενο βράδυ. Αρχικά
το έντομο βρίσκεται στο μπάνιο. Αν {X n } n∈N 0 είναι η μαρκοβιανή αλυσίδα
στον χώρο καταστάσεων X ={Κ, Β, Σ, Υ, Μ} που περιγράφει τη θέση του
εντόμου μετά από n βράδυα, βρείτε τον πίνακα πιθανοτήτων μετάβασής της.

Αρχικά κατασκευάζουμε τον πίνακα μεταβάσεων και την αρχική κατανομή.

### Πίνακας πιθανοτήτων μετάβασης (markov_table)

```python
from simple_markov_chain_lib import markov_chain  # import markov chain simulator

# Transition Table
markov_table = {
    'Κ': {'Β': 1/2, 'Σ': 1/2},    # Από την κατάσταση Κ, μεταφέρομαι στις δυνατές καταστάσεις ισοπίθανα.
    'Β': {'Κ': 1/2, 'Σ': 1/2},    # Αντίστοιχα για τις υπόλοιπες
    'Σ': {'Β': 1/4, 'Κ': 1/4, 'Υ':1/4,'Μ':1/4},
    'Υ': {'Σ': 1},
    'Μ': {'Σ':1}
}
```

### Αρχική κατανομή 

(Εδώ υποθέτουμε ότι την πρώτη μέρα, το έντομο βρίσκεται στο μπάνιο, όπως ορίζει η εκφώνηση)

```python
# Initial Distribution
init_dist = {'Μ': 1}  # Ξεκινάμε από την κατάσταση Μ με πιθανότητα 1 
```

### Δημιουργία Νέας Markov Chain με την αρχική κατανομή που ορίστηκε άνωθεν

```python
mc = markov_chain(markov_table, init_dist)
```

### Περίπατος 20 βημάτων (1η εώς 21η μέρα):

Τρέχουμε τον πίνακα μεταβάσεων για 20 βήματα (1η εώς 21η ημέρα) και κάνουμε print την κατάσταση σε κάθε βήμα.

```python
mc.start()
print("Το βράδυ της {}ης μέρας, το έντομο βρίσκεται στο δωμάτιο {}".format(mc.steps+1, mc.running_state))   
for i in range(20):
    mc.move()
    print("Το βράδυ της {}ης μέρας, το έντομο βρίσκεται στο δωμάτιο {}".format(mc.steps+1, mc.running_state))
   
```
